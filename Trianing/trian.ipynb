```python
# รวมทั้ง 3 ไฟล์เป็น DataFrame เดียว
files = ['202604101208.csv', '202604101240.csv', '202604101246.csv']

df = pd.concat([pd.read_csv(f'data/{f}') for f in files], ignore_index=True)

print(f"✅ รวมข้อมูลสำเร็จ")
print(f"จำนวนข้อมูลทั้งหมด: {df.shape[0]:,} แถว")
print(f"คอลัมน์: {df.shape[1]} คอลัมน์")
print()
print(df.head())
```

```python
# ดูว่าแต่ละคอลัมน์มี NaN เท่าไหร่
print("=== ค่า NaN ในแต่ละคอลัมน์ ===")
print(df.isnull().sum())

print("\n=== ตัวอย่าง treatment ที่มีในข้อมูล ===")
print(df['treatment'].dropna().unique()[:20])

print("\n=== ช่วงเวลาของข้อมูล ===")
print("เริ่มต้น:", df['appointment_start'].min())
print("สิ้นสุด:", df['appointment_start'].max())
```

```python
# === Step 1: กรองแถวที่จำเป็นต้องมีข้อมูล ===
df_clean = df.copy()

df_clean = df_clean.dropna(subset=['receipt_time', 'check_in_time', 'treatment'])

print(f"หลังกรอง NaN: {len(df_clean):,} แถว")

# === Step 2: แปลงเวลาเป็น datetime ===
df_clean['receipt_time'] = pd.to_datetime(df_clean['receipt_time'])
df_clean['check_in_time'] = pd.to_datetime(df_clean['check_in_time'])
df_clean['appointment_start'] = pd.to_datetime(df_clean['appointment_start'])

# === Step 3: คำนวณ actual duration ===
df_clean['actual_duration_min'] = (
    df_clean['receipt_time'] - df_clean['check_in_time']
).dt.total_seconds() / 60

# กรองค่าผิดปกติออก
df_clean = df_clean[
    (df_clean['actual_duration_min'] >= 10) &
    (df_clean['actual_duration_min'] <= 150)
]

print(f"หลังกรอง duration ผิดปกติ: {len(df_clean):,} แถว")

# === Step 4: ดู distribution ===
print("\n=== การกระจายของ duration (นาที) ===")
print(df_clean['actual_duration_min'].describe())
```

```python
# === สร้าง Label 7 ระดับ ===
def assign_duration_class(minutes):
    if minutes <= 22:    return 15
    elif minutes <= 37:  return 30
    elif minutes <= 52:  return 45
    elif minutes <= 67:  return 60
    elif minutes <= 82:  return 75
    elif minutes <= 97:  return 90
    else:                return 105

df_clean['duration_class'] = df_clean['actual_duration_min'].apply(assign_duration_class)

# ดูว่าแต่ละ class มีข้อมูลเท่าไหร่
print("=== จำนวนข้อมูลในแต่ละ class ===")
counts = df_clean['duration_class'].value_counts().sort_index()
for cls, cnt in counts.items():
    pct = cnt / len(df_clean) * 100
    print(f"  {cls:3d} นาที : {cnt:6,} แถว ({pct:.1f}%)")
```

```python
# === Feature Engineering ===

# 1. Tooth Count — นับจำนวนซี่ฟัน
df_clean['tooth_count'] = df_clean['tooth_no'].apply(
    lambda x: len(str(x).split(',')) if pd.notna(x) else 1
)

# 2. Treatment Count — นับจำนวนหัตถการ
df_clean['treatment_count'] = df_clean['treatment'].apply(
    lambda x: len(str(x).split(',')) if pd.notna(x) else 1
)

# 3. Has Complex Treatment — ตรวจหาหัตถการซับซ้อน
COMPLEX_KEYWORDS = [
    'root', 'endo', 'รากฟัน', 'crown', 'ครอบ', 'bridge', 'สะพาน',
    'implant', 'ฝัง', 'extract', 'ถอน', 'ผ่าตัด', 'surgery'
]

def has_complex(text):
    if pd.isna(text):
        return 0
    text_lower = str(text).lower()
    return int(any(kw in text_lower for kw in COMPLEX_KEYWORDS))

df_clean['has_complex_treatment'] = df_clean['treatment'].apply(has_complex)

# 4. Time of Day
df_clean['hour'] = df_clean['check_in_time'].dt.hour

def get_time_of_day(hour):
    if 6 <= hour < 12:   return 0  # morning
    elif 12 <= hour < 17: return 1  # afternoon
    else:                 return 2  # evening

df_clean['time_of_day_enc'] = df_clean['hour'].apply(get_time_of_day)

# 5. Is First Case ของวันนั้น
df_clean['date'] = df_clean['check_in_time'].dt.date
df_clean = df_clean.sort_values(['branch_id', 'check_in_time'])
df_clean['is_first_case'] = (
    df_clean.groupby(['branch_id', 'date']).cumcount() == 0
).astype(int)

# 6. Branch Median Duration (แทน doctor profile เพราะ license_no หายเยอะ)
branch_stats = df_clean.groupby('branch_id')['actual_duration_min'].median()
GLOBAL_MEDIAN = df_clean['actual_duration_min'].median()
df_clean['branch_median_duration'] = df_clean['branch_id'].map(branch_stats).fillna(GLOBAL_MEDIAN)

# สรุป features ที่สร้างได้
FEATURE_COLS = [
    'treatment_count',
    'has_complex_treatment', 
    'tooth_count',
    'time_of_day_enc',
    'is_first_case',
    'branch_median_duration'
]

print("✅ Feature Engineering สำเร็จ")
print(f"\nFeatures ที่ใช้: {FEATURE_COLS}")
print(f"\nตัวอย่างข้อมูล:")
print(df_clean[FEATURE_COLS + ['duration_class']].head(10))
```

```python
# === แบ่งข้อมูล Train / Val / Test แบบ Chronological ===
df_clean['month'] = df_clean['check_in_time'].dt.month
df_clean['day']   = df_clean['check_in_time'].dt.day

train_df = df_clean[df_clean['month'] <= 3]
val_df   = df_clean[(df_clean['month'] == 4) & (df_clean['day'] <= 15)]
test_df  = df_clean[(df_clean['month'] == 4) & (df_clean['day'] > 15)]

X_train = train_df[FEATURE_COLS]
y_train = train_df['duration_class']

X_val   = val_df[FEATURE_COLS]
y_val   = val_df['duration_class']

X_test  = test_df[FEATURE_COLS]
y_test  = test_df['duration_class']

print("=== จำนวนข้อมูลในแต่ละ set ===")
print(f"Train : {len(X_train):,} แถว")
print(f"Val   : {len(X_val):,} แถว")
print(f"Test  : {len(X_test):,} แถว")
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

# === เตรียม Label Encoder ===
# XGBoost ต้องการ label เป็น 0-6 ไม่ใช่ 15,30,45...
le = LabelEncoder()
le.fit([15, 30, 45, 60, 75, 90, 105])

y_train_enc = le.transform(y_train)
y_val_enc   = le.transform(y_val)
y_test_enc  = le.transform(y_test)

print("Class mapping:")
for i, c in enumerate(le.classes_):
    print(f"  index {i} → {c} นาที")

# === Train XGBoost ===
print("\nกำลัง Train โมเดล...")

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    scale_pos_weight=1,
    eval_metric='mlogloss',
    random_state=42,
    verbosity=0
)

model.fit(
    X_train, y_train_enc,
    eval_set=[(X_val, y_val_enc)],
    verbose=50
)

print("\n✅ Train เสร็จแล้ว")
```

```python
from sklearn.metrics import f1_score, mean_absolute_error, classification_report

# === คำนวณ Metrics ===
y_pred_enc = model.predict(X_test)
y_pred     = le.inverse_transform(y_pred_enc)  # แปลงกลับเป็น minutes
y_test_min = y_test.values

# Macro F1
macro_f1 = f1_score(y_test_min, y_pred, average='macro')

# MAE
mae = mean_absolute_error(y_test_min, y_pred)

# Under-estimation Rate
under_est_rate = np.mean(y_pred < y_test_min)

print("=== ผลการ Evaluate บน Test Set ===")
print(f"Macro F1-score       : {macro_f1:.4f}")
print(f"MAE (minutes)        : {mae:.2f}")
print(f"Under-estimation Rate: {under_est_rate:.4f}")

print("\n=== Classification Report ===")
print(classification_report(y_test_min, y_pred))
```

```python
# === เพิ่ม Features ที่ดีขึ้น ===

# 1. Duration จาก appointment ที่นัดไว้ (scheduled duration)
df_clean['appointment_end'] = pd.to_datetime(df_clean['appointment_end'])
df_clean['scheduled_duration'] = (
    df_clean['appointment_end'] - df_clean['appointment_start']
).dt.total_seconds() / 60

# 2. Total amount — ราคาสูง มักสัมพันธ์กับหัตถการซับซ้อน
df_clean['total_amount'] = pd.to_numeric(df_clean['total_amount'], errors='coerce').fillna(0)

# 3. Has notes — มี notes มักหมายถึงเคสซับซ้อน
df_clean['has_notes'] = df_clean['notes'].notna().astype(int)

# 4. Day of week — วันจันทร์อาจยุ่งกว่าวันศุกร์
df_clean['day_of_week'] = df_clean['check_in_time'].dt.dayofweek

# อัปเดต FEATURE_COLS
FEATURE_COLS = [
    'treatment_count',
    'has_complex_treatment',
    'tooth_count',
    'time_of_day_enc',
    'is_first_case',
    'branch_median_duration',
    'scheduled_duration',   # ← สำคัญมาก
    'total_amount',
    'has_notes',
    'day_of_week'
]

# กรอง scheduled_duration ที่ผิดปกติออก
df_clean = df_clean[
    (df_clean['scheduled_duration'] >= 5) &
    (df_clean['scheduled_duration'] <= 180)
]

print(f"ข้อมูลหลังเพิ่ม features: {len(df_clean):,} แถว")
print(f"\nFeatures ทั้งหมด: {FEATURE_COLS}")
print(f"\nตัวอย่าง scheduled_duration:")
print(df_clean['scheduled_duration'].describe())
```

```python
# === แบ่งข้อมูลใหม่ ===
train_df = df_clean[df_clean['month'] <= 3]
val_df   = df_clean[(df_clean['month'] == 4) & (df_clean['day'] <= 15)]
test_df  = df_clean[(df_clean['month'] == 4) & (df_clean['day'] > 15)]

X_train = train_df[FEATURE_COLS]
y_train = train_df['duration_class']
X_val   = val_df[FEATURE_COLS]
y_val   = val_df['duration_class']
X_test  = test_df[FEATURE_COLS]
y_test  = test_df['duration_class']

y_train_enc = le.transform(y_train)
y_val_enc   = le.transform(y_val)
y_test_enc  = le.transform(y_test)

print(f"Train: {len(X_train):,} | Val: {len(X_val):,} | Test: {len(X_test):,}")

# === Train ใหม่ ===
print("\nกำลัง Train โมเดลใหม่...")

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    eval_metric='mlogloss',
    random_state=42,
    verbosity=0
)

model.fit(
    X_train, y_train_enc,
    eval_set=[(X_val, y_val_enc)],
    verbose=50
)

# === Evaluate ===
y_pred_enc = model.predict(X_test)
y_pred     = le.inverse_transform(y_pred_enc)
y_test_min = y_test.values

macro_f1       = f1_score(y_test_min, y_pred, average='macro')
mae            = mean_absolute_error(y_test_min, y_pred)
under_est_rate = np.mean(y_pred < y_test_min)

print("\n=== ผลการ Evaluate ===")
print(f"Macro F1-score       : {macro_f1:.4f}")
print(f"MAE (minutes)        : {mae:.2f}")
print(f"Under-estimation Rate: {under_est_rate:.4f}")
print("\n=== Classification Report ===")
print(classification_report(y_test_min, y_pred))
```

```python
import pandas as pd

train_df = pd.read_parquet('data/features_train.parquet')
test_df  = pd.read_parquet('data/features_test.parquet')

print("Train:", train_df.shape)
print("Test:", test_df.shape)
print("\nคอลัมน์ทั้งหมด:")
print(train_df.columns.tolist())
print("\nLabel distribution:")
print(train_df['duration_class'].value_counts().sort_index())
```

```python
import numpy as np
import joblib
import json
import os
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.metrics import f1_score, mean_absolute_error, classification_report

# === Features จริงจาก Sun ===
FEATURE_COLS = [
    'treatment_class', 'composite_treatment_flag', 'has_tooth_no',
    'tooth_count', 'is_area_treatment', 'surface_count', 'total_amount',
    'has_notes', 'appt_day_of_week', 'appt_hour_bucket', 'is_first_case',
    'has_dentist_id', 'appointment_rank_in_day', 'clinic_median_duration',
    'clinic_pct_long', 'doctor_median_duration', 'doctor_pct_long',
    'scheduled_duration_min'
]

X_train = train_df[FEATURE_COLS]
y_train = train_df['duration_class']
X_test  = test_df[FEATURE_COLS]
y_test  = test_df['duration_class']

# === Label Encoder (6 class ไม่มี 75) ===
le = LabelEncoder()
le.fit([15, 30, 45, 60, 90, 105])

y_train_enc = le.transform(y_train)
y_test_enc  = le.transform(y_test)

print("Class mapping:")
for i, c in enumerate(le.classes_):
    print(f"  index {i} → {c} นาที")

# === Sample weights ช่วย minority class ===
sample_weights = compute_sample_weight(
    class_weight='balanced',
    y=y_train_enc
)

# === Train ===
print("\nกำลัง Train โมเดล...")

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    eval_metric='mlogloss',
    random_state=42,
    verbosity=0,
    n_jobs=-1
)

model.fit(
    X_train, y_train_enc,
    sample_weight=sample_weights,
    verbose=50
)

# === Evaluate ===
y_pred_enc = model.predict(X_test)
y_pred     = le.inverse_transform(y_pred_enc)
y_test_min = y_test.values

macro_f1       = f1_score(y_test_min, y_pred, average='macro')
mae            = mean_absolute_error(y_test_min, y_pred)
under_est_rate = np.mean(y_pred < y_test_min)

print("\n=== ผลการ Evaluate ===")
print(f"Macro F1-score       : {macro_f1:.4f}")
print(f"MAE (minutes)        : {mae:.2f}")
print(f"Under-estimation Rate: {under_est_rate:.4f}")
print("\n=== Classification Report ===")
print(classification_report(y_test_min, y_pred))
```
